##A célula de codigo a seguir está gerando 200 padrões aleatórios para uma distribuição normal multivariada.

##$$X \sim N(\mu_x, \Sigma) $$

##A matriz X corresponde a uma amostra contendo 200 padrões da distribuição normal multivariada acima.

```python
import numpy as np
import matplotlib.pyplot as plt
mu_x=[0, 0]
Sigma=[[1.1,2.1],[2.1,5]]

X = np.random.multivariate_normal(mu_x, Sigma, 200)

plt.figure(figsize = (10,10))
plt.plot(X[:,0], X[:,1],'xb')
plt.axis('equal')
plt.show()
```

##O código a seguir estima os valores do vetor de média e da matriz de covariância a partir da amostra X note a diferença em relação aos valores originais.

```python
S_x = np.cov(X.T)
m_x = np.mean(X,axis=0)

print('Media amostral\n', m_x,'\n')
print('Matriz de covariância amostral\n',S_x,'\n')
```

##**Questão 1:** Gere na célula de código abaixo as seguintes quantidades de padrões (5, 50, 500, 5.000 e 50.000). Avalie numericamente a qualidade da estimação de $ \mu $ e $ \Sigma $ obtidas a medida que varia o número padrões na amosta. Utilize neste experimeto $\mu = \{ 0,0\}$ e $\Sigma \begin{bmatrix}
1.1 & 2.1 \\
2.1 & 5
\end{bmatrix}$.

```python
mu_x=[0, 0]
Sigma=[[1.1,2.1],[2.1,5]]
patterns_n = [5, 50, 500, 5000, 50000]
sigmas = []
mu = []
plt.figure(figsize = (20,20))

for i, patterns in enumerate(patterns_n):
  X = np.random.multivariate_normal(mu_x, Sigma, patterns)
  S_x = np.cov(X.T)
  m_x = np.mean(X,axis=0)

  sigmas.append(np.linalg.norm(S_x-Sigma))
  mu.append(np.linalg.norm(m_x-mu_x))

  print(f'norma da Media amostral para {patterns} padrões => ', np.linalg.norm(m_x-mu_x) ,'')
  print(f'norma da Matriz de covariância amostral para {patterns} padrões => ',np.linalg.norm(S_x-Sigma),'\n\n')

  plt.subplot(5,5,i+1)
  plt.plot(X[:,0], X[:,1],'xb')
  plt.title(f"{patterns} padrões")
  plt.axis('equal')

print(f"menor mu: {patterns_n[np.argmin(mu)]}")
print(f"menor sigma: {patterns_n[np.argmin(sigmas)]}")

plt.tight_layout(pad=3.0)
plt.show()



```

```python
plt.plot(mu, color='purple', label='mu')
plt.plot(sigmas, color='orange', label='sigmas')

# Add labels and title
plt.title(f'Normas da diferença entre mu obtidos \n e original e sigmas obtivos e original')

# Add legend
plt.legend()

# Display the plot
plt.show()
```

### Comentários:
1. Para a Execução desse experimento foi utilizada a norma euclidiana da diferença entre os valores médios e matrizes de covariância experimentais e aqueles definidos previamente.

2. Como pode ser observado no gráfico, quanto maior a quantidade de padrões gerados, mais próximo de zero é a diferença entre os valores experimentais de $\mu$ e $\Sigma$ e aqueles utilizados para gerar a distribuição Normal Multivariada.

##A célula de codigo a seguir está gerando 200 padrões aleatórios para duas
distribuições normais multivariadas.

##$$X \sim N(\mu_x, \Sigma)\quad \textrm{e} \quad Y \sim N(\mu_y, \Sigma)$$

## Na célula de código a seguir as matrizes X (em preto no gráfico) e Y (apresentada em vermelho no gráfico) correspondem a duas amostras contendo 200 padrões cada das distribuições normais multivariada acima. Observe estas amostras no gráfico abaixo.

```python
import numpy as np
import matplotlib.pyplot as plt
mu_x=[0, 0]
mu_y=[4,3]
Sigma=[[1.1,2.1],[2.1,5]]

X = np.random.multivariate_normal(mu_x, Sigma, 200)
Y = np.random.multivariate_normal(mu_y, Sigma, 200)

plt.figure(figsize = (10,10))
plt.plot(X[:,0], X[:,1],'xk')
plt.plot(Y[:,0], Y[:,1],'xr')
plt.axis('equal')
plt.show()
```

#**Questão 2**: Estime os valores de $ \mu $ e $ \Sigma $ para as amostras X e Y. Em seguida, utilizando probabilidades a priori iguais para ambas as classes, insira no gráfico acima as fronteiras de decisão obtidas para o classificador Bayesiano para os valores estimados e para os valores verdadeiros.   

```python
S_x = np.cov(X.T)
S_y = np.cov(Y.T)
me_x = np.mean(X,axis=0)
me_y = np.mean(Y,axis=0)

#implementação da função que define a fronteira de decisão:
def fronteira(X, mu1, mu2, sigma1, sigma2):
  term1 = (X - mu1).T @ np.linalg.inv(sigma1) @ (X - mu1)
  term2 = (X - mu2).T @ np.linalg.inv(sigma2) @ (X - mu2)
  return term1 - term2

#definição do espaço para calculo da fronteira
dimensaox = np.linspace(-5, 10, 400)
dimensaoy = np.linspace(-5, 10, 400)
X_grid, Y_grid = np.meshgrid(dimensaox, dimensaoy)
points = np.c_[X_grid.ravel(), Y_grid.ravel()]

#cálculo das fronteiras e ajuste dos arrays retornados para serem plotados na função contour
fronteira_est = np.array([fronteira(point, me_x, me_y, S_x, S_y) for point in points])
fronteira_est = fronteira_est.reshape(X_grid.shape)
fronteira_real = np.array([fronteira(point, mu_x, mu_y, Sigma, Sigma) for point in points])
fronteira_real = fronteira_real.reshape(X_grid.shape)


plt.figure(figsize = (10,10))

plt.plot(X[:,0], X[:,1],'xk', label='Classe X')
plt.plot(Y[:,0], Y[:,1],'xr', label='Classe Y')
plt.contour(X_grid, Y_grid, fronteira_est, levels=[0], color='cyan', linestyles='dashed', label='fronteira de decisão')
plt.contour(X_grid, Y_grid, fronteira_real, levels=[0], colors='magenta', linestyles='dotted')
plt.axis('equal')
plt.legend()
plt.show()

'''
1. fronteira é a função que calcula a fronteira de decisão
2. o plot é obtido a partir da função contour que exibe no
nível onde o resultado da função é zero
3. a fronteira de decisão para os valores reais foi calculada usando mu_x, mu_y e Sigma
'''
```

```python
plt.figure(figsize = (20,20))

plt.subplot(3,3,1)
plt.plot(X[:,0], X[:,1],'xk', label='Classe X')
plt.plot(Y[:,0], Y[:,1],'xr', label='Classe Y')
plt.contour(X_grid, Y_grid, fronteira_est, levels=[0], color='cyan', linestyles='dashed', label='fronteira de decisão')
plt.contour(X_grid, Y_grid, fronteira_real, levels=[0], colors='magenta', linestyles='dotted')
plt.axis('equal')
plt.title('fronteiras sobrepostas')
plt.legend()

plt.subplot(3,3,2)
plt.plot(X[:,0], X[:,1],'xk')
plt.plot(Y[:,0], Y[:,1],'xr')
plt.contour(X_grid, Y_grid, fronteira_est, levels=[0], colors='cyan', linestyles='dotted') # plot no espaço de X e Y _grid onde os termos de fronteira_est são iguais a zero, ou seja, a fronteira desejada
plt.axis('equal')
plt.title('fronteira estimada')

plt.subplot(3,3,3)
plt.plot(X[:,0], X[:,1],'xk')
plt.plot(Y[:,0], Y[:,1],'xr')
plt.contour(X_grid, Y_grid, fronteira_real, levels=[0], colors='magenta', linestyles='dotted') # plot no espaço de X e Y _grid onde os termos de fronteira_est são iguais a zero, ou seja, a fronteira desejada
plt.axis('equal')
plt.title('fronteira real')


plt.show()
```

### Comentários:
1. Os resultados obtidos foram bastante próximos e praticamente se sobrepuserem em grande parte dos pontos.
2. A separação gerada por ambas as fronteiras foi bastante satisfatória, pois apenas um padrão da classe Y estava na área da classe X.
3. O valor X mais superior, o qual é membro da classe X, seria melhor identificado pela fronteira estimada, Dessa forma, as duas fronteiras não são 100% exatas, porém possuem altíssimo grau de acerto, logo, são confiáveis para separar esse conjunto de dados.

#**Questão 3**: Trace em um gráfico as fronteiras de decisão para o problema de duas classes anterior usando os valores verdadeiros de $\mu$ e $\Sigma$ variando os valores das probabilidades a priori, trace as fronteiras de decisão para a probabilidade a priori $P(X)$ igual a $\{0.2, 0.4, 0.6 \, \textrm{e} \, 0.8\}$.
PS.: Não esqueça de variar a probabilidade a priori de um objeto pretencer à classe Y de forma coerente.

```python
apriori_1 = [0.2, 0.4, 0.6, 0.8]
fronteiras_dif = []
def fronteira_dif(X, mu1, mu2, sigma1, sigma2, pi1):
    pi2 = 1 - pi1
    return fronteira(X, mu1, mu2, sigma1, sigma2) - 2 * np.log(pi2 / pi1)

dimensaox = np.linspace(-5, 13, 400)
dimensaoy = np.linspace(-5, 13, 400)
X_grid, Y_grid = np.meshgrid(dimensaox, dimensaoy)
points = np.c_[X_grid.ravel(), Y_grid.ravel()]

for ap1 in apriori_1:
  fronteira_est = np.array([fronteira_dif(point, me_x, me_y, S_x, S_y,ap1) for point in points])
  fronteiras_dif.append(fronteira_est.reshape(X_grid.shape))

cor=['black','orange','magenta','green']
plt.figure(figsize = (25,25))

for i in range(len(apriori_1)):
  plt.subplot(2,2,i+1)
  plt.plot(X[:,0], X[:,1],'xk')
  plt.plot(Y[:,0], Y[:,1],'xr')
  plt.contour(X_grid, Y_grid, fronteiras_dif[i], levels=[0], colors=cor[i], linestyles='dotted')
  plt.axis('equal')
  plt.title(f'fronteira prob apriori1 = {apriori_1[i]}')




```

```python
cor=['black','orange','magenta','green']
plt.figure(figsize = (20,20))

plt.subplot(2,2,i+1)
plt.plot(X[:,0], X[:,1],'xk')
plt.plot(Y[:,0], Y[:,1],'xr')
for i in range(len(apriori_1)):
  plt.contour(X_grid, Y_grid, fronteiras_dif[i], levels=[0], colors=cor[i], linestyles='dotted')
plt.axis('equal')
plt.title(f'fronteiras sobrepostas')


```

1. A medida que a probabilidade a priori de X cresce, a fronteira de decisão se aproxima do conjunto X, uma vez que é fornecido ao problema ser mais esperado encontrar valores da classe X.
2. Diante disso, a fronteira será "mais exata" nas proximidades da classe mais esperada, ou seja, mais próxima desse conjunto do que do outro. Por isso o deslocamento
  
  2.1 Como pode ser observado pelas fronteiras verde e preta; aquela mais próxima de X, pois $P(X) = 0.8$, e essa mais próxima de Y, pois $P(X)=0.2$, logo, $P(Y) = 0.8$

##**Questão 4**: Projete as amostras X e Y na primeira PCA. Obtenha a fronteira de decisão no espaço da PCA para o classificador Bayesiano utilizando probabilidades a priori iquais a 0.5. Apresente seus resultados graficamente, utilizando a célula de código abaixo.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

#normalização dos dados
scaler = StandardScaler()
Xnorm = scaler.fit_transform(X)
Ynorm = scaler.transform(Y)

#aplicação do pca
pca = PCA(n_components=2)
X_pca = pca.fit_transform(Xnorm)
Y_pca = pca.transform(Ynorm)
```

```python
#calculo das matrizes de covariância e mu
S_x = np.cov(X_pca.T)
S_y = np.cov(Y_pca.T)
me_x = np.mean(X_pca,axis=0)
me_y = np.mean(Y_pca,axis=0)

#calculo das fronteiras de decisão
dimensaox = np.linspace(-3, 8, 800)
dimensaoy = np.linspace(-3, 8, 800)
X_grid, Y_grid = np.meshgrid(dimensaox, dimensaoy)
points = np.c_[X_grid.ravel(), Y_grid.ravel()]

fronteira_real = np.array([fronteira(point, me_x, me_y, S_x, S_y) for point in points])
fronteira_real = fronteira_real.reshape(X_grid.shape)
```

```python
plt.figure(figsize=(20, 20))

plt.subplot(3, 3, 1)
plt.scatter(X[:, 0], X[:, 1],color='black', alpha=0.5)
plt.scatter(Y[:, 0], Y[:, 1],color='red', alpha=0.5)
plt.title('Original Data')

plt.subplot(3, 3, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1],color='black', alpha=0.5)
plt.scatter(Y_pca[:, 0], Y_pca[:, 1],color='red', alpha=0.5)
plt.title('Conjunto após o PCA')

plt.subplot(3, 3, 3)
plt.scatter(X_pca[:,0], X_pca[:,1],color='black', alpha=0.3, label='Classe X')
plt.scatter(Y_pca[:,0], Y_pca[:,1],color='red', alpha=0.3, label='Classe Y')
plt.contour(X_grid, Y_grid, fronteira_real, levels=[0], colors='magenta', linestyles='dashed')
plt.axis('equal')
plt.title('fronteiras de decisao')

plt.legend()
plt.tight_layout()
plt.show()
```

### Obs:
I. Os dados foram normalizados para que o PCA fosse calculado

II. A preparação, representada pelo termo "*fit*" foi realizada apenas para a classe X. Para a classe Y foi realizada a transformação, assim como para X, mas sem o "treinamento" do pca.

III. A partir dos valores do pca para X e Y, foram calculadas as fronteiras para probabilidades a priori iguais

---
### Comentários:
1. A fronteira ficou muito mais clara de ser visualizada e a separação dos dados tornou-se melhor, porém sem a perda da variabilidade das informações, como o pca se propõe a fazer.
2. Ainda assim, um valor da classe Y não foi colocado no lado certo da fronteira, entretanto, a confiabilidade da fronteira definida através dos dados pós pca é ala.
